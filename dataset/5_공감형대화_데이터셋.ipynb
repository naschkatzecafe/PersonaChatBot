```python
# 압축 푼 파일에 대해 정서와 관계 파악하기
import os
from collections import Counter

# 압축 해제된 데이터가 있는 폴더 경로
data_path = '/content/drive/MyDrive/extracted_data'

# 파일 목록 읽기
files = os.listdir(data_path)

# 정서와 관계를 저장할 카운터
emotion_counter = Counter()
relationship_counter = Counter()

# 각 파일명에서 정서와 관계 추출 및 카운트
for file in files:
    if file.startswith('Empathy_') and file.endswith('.json'):
        parts = file.split('_')
        if len(parts) >= 3:
            emotion = parts[1]
            relationship = parts[2]
            emotion_counter[emotion] += 1
            relationship_counter[relationship] += 1

# 결과 출력
print("정서 카운트:")
for emotion, count in emotion_counter.items():
    print(f"{emotion}: {count}")

print("\n관계 카운트:")
for relationship, count in relationship_counter.items():
    print(f"{relationship}: {count}")
```

```python
#관계별 정서 카운트
import os
from collections import defaultdict, Counter

# 압축 해제된 데이터가 있는 폴더 경로
data_path = '/content/drive/MyDrive/extracted_data'

# 파일 목록 읽기
files = os.listdir(data_path)

# 관계별로 각 정서를 카운트할 defaultdict 생성
relationship_emotion_counter = defaultdict(Counter)

# 각 파일명에서 정서와 관계 추출 및 카운트
for file in files:
    if file.startswith('Empathy_') and file.endswith('.json'):
        parts = file.split('_')
        if len(parts) >= 3:
            emotion = parts[1]
            relationship = parts[2]
            relationship_emotion_counter[relationship][emotion] += 1

# 결과 출력
print("관계별 각 정서 카운트:")
for relationship, emotions in relationship_emotion_counter.items():
    print(f"{relationship}:")
    for emotion, count in emotions.items():
        print(f"  {emotion}: {count}")
```

```python
#그래프
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=len(relationship_emotion_counter), figsize=(12, 4 * len(relationship_emotion_counter)))
if len(relationship_emotion_counter) == 1:
    axes = [axes]  # subplot이 하나일 경우 axes를 리스트로 만들어 일관된 처리를 가능하게 함

for ax, (relationship, emotions) in zip(axes, relationship_emotion_counter.items()):
    # 각 관계에 대한 정서별 빈도
    emotions = dict(emotions)
    emotion_names = list(emotions.keys())
    emotion_counts = list(emotions.values())

    # 막대그래프 그리기
    ax.bar(emotion_names, emotion_counts, color='skyblue')
    ax.set_title(f'정서 빈도수: {relationship}')
    ax.set_xlabel('정서')
    ax.set_ylabel('빈도수')

# 레이아웃 조정
plt.tight_layout()
plt.show()
```

모든 파일 A, B 주고받는 대화 + 감정으로 데이터셋 준비하기

```python
import os
import json
import pandas as pd

# 파일 목록을 읽어올 경로
extract_path = '/content/drive/MyDrive/extracted_data'
files = os.listdir(extract_path)
print(len(files))
```

```python
# 데이터(대화) 당 발화들 쪼개기
def process_utterances(utterances, info):
    results = []
    i = 0
    while i < len(utterances) - 1:
        current = utterances[i]
        next_utter = utterances[i + 1]
        # print('current: ', current)
        # print('next_current', next_utter)

        if current['speaker_changeEmotion'] is None:
            combined_text = current['text'] + " " + next_utter['text']
            # print('combined: ', combined_text)
            emotion = info['speaker_emotion']
            results.append({'text': combined_text, 'emotion': emotion})
            i += 1
        else:
            results.append({'text': current['text'], 'emotion': current['speaker_changeEmotion']})
            i += 1
    return results
```

```python
# 전체 결과를 저장할 데이터 프레임
import time

all_results = []

# 테스트용.
# files = files[:1]

for file in files:
  if file.endswith('.json'):
    file_path = os.path.join(extract_path, file)
    with open(file_path, 'r', encoding='utf-8') as f:
      data = json.load(f)
      info = data['info']
      utterances = data['utterances']
      s = time.time()
      file_results = process_utterances(utterances, info)
      e = time.time()
      print(f'{e-s} sec, {file_path}')
      all_results.extend(file_results)

```

```python
# 모든 파일에서 추출된 데이터로 데이터 프레임 생성
df = pd.DataFrame(all_results)

print(df)
```
