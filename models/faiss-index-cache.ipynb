## 대화 데이터

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Pal-ette/preprocessed_data/emotion.csv')
df
```

```python
result_df = pd.DataFrame(columns=['query', 'answer'])
result_df = pd.concat([result_df, pd.DataFrame({'query': df['사람문장1'], 'answer': df['시스템문장1']})], ignore_index=True)
result_df = pd.concat([result_df, pd.DataFrame({'query': df[['사람문장1', '시스템문장1', '사람문장2']].agg(' '.join, axis=1), 'answer': df['시스템문장2']})], ignore_index=True)
result_df
```

## 임베딩

```python
from sentence_transformers import SentenceTransformer
from tqdm.auto import tqdm
import numpy as np

model = SentenceTransformer('jhgan/ko-sroberta-multitask')
embeddings = [None] * result_df.shape[0]

for idx, (query, _answer) in tqdm(result_df.iterrows(), desc="Embedding...", total=result_df.shape[0]):
    embedding = model.encode(query)
    embeddings[idx] = embedding
embeddings = np.array(embeddings)
```

## FAISS 인덱싱

```python
import faiss

index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(embeddings)
```

## 검색 테스트

```python
def search(query, index, k=1):
    query_embedding = model.encode(query)
    distances, indices = index.search(query_embedding.reshape(1, -1), k)
    return result_df.iloc[indices[0][0]]['answer']
```

```python
search("오늘 오는 길에 귀여운 강아지를 만났어", index)
```

```python
search("로또에 당첨됐어!", index)
```

```python
search("일은 왜 해도 해도 끝이 없을까? 화가 난다.", index)
```

## 인덱스와 데이터 파일로 저장

```python
faiss.write_index(index, "/content/drive/MyDrive/Pal-ette/trained_models/embedding_index.index")
result_df.to_csv("/content/drive/MyDrive/Pal-ette/trained_models/embedding_data.csv")
```
